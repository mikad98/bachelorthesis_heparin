### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pandas.io import gbq
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures

pd.options.mode.chained_assignment = None  # default='warn'
```

### APTT Data

```python
# Load Data
query = """
        SELECT *
        FROM `bachelorarbeit-heparin.mimic_data.cohort1_final`
        """
df = gbq.read_gbq(query, project_id = "bachelorarbeit-heparin")

# Plot the data:
plt.figure(figsize=(15,10))
plt.xlabel('Time [h] after initial Heparin Treatment', fontsize=32)
plt.ylabel('APTT [sec]', fontsize=32)
plt.ylim(0,160)
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)
plt.scatter(df['rel_charttime']/60, df['PTT'])
```

### Correlation of Administration Rate and Patientweight

```python
# Load Data
query = """
        SELECT *
        FROM `bachelorarbeit-heparin.mimic_data.cohort1_final`
        """
df = gbq.read_gbq(query, project_id = "bachelorarbeit-heparin")

# Plot the data:
plt.figure(figsize=(15,10))
plt.xlabel('Patientweight [kg]', fontsize=32)
plt.ylabel('Heparin Administratin Rate [units]', fontsize=32)
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)
plt.scatter(df['patientweight'], df['rate_by_weight']*df['patientweight'])
```
